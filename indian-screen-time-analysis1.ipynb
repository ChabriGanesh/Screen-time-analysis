```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
df = pd.read_csv(r"C:\Users\Chabri Ganesh\Indian_Kids_Screen_Time.csv")  
# Display first few rows
print("ðŸ“Š Data Preview:")
print(df.head())
# Summary statistics
print("\nðŸ“ˆ Summary Stats:")
print(df.describe())
```

```python
# Average screen time per age group
age_group_mean = df.groupby("Age")["Avg_Daily_Screen_Time_hr"].mean()
print("\nAverage Screen Time by Age (hrs/day):")
print(age_group_mean)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=age_group_mean.index,
    y=age_group_mean.values,
    hue=age_group_mean.index,  # Needed for palette
    palette="viridis",
    legend=False
)
plt.title("Average Screen Time by Age")
plt.xlabel("Age")
plt.ylabel("Screen Time (hours/day)")
plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Gender",
    y="Avg_Daily_Screen_Time_hr",
    hue="Gender",           # For palette colors per category
    data=df,
    palette="pastel",
    legend=False
)
plt.title("Screen Time Distribution by Gender")
plt.show()
```

```python
# Distribution plot
plt.figure(figsize=(6,4))
sns.histplot(df["Avg_Daily_Screen_Time_hr"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Screen Time (Hours)")
plt.xlabel("Hours per Day")
plt.ylabel("Count")
plt.show()
```

```python
# Urban vs Rural - Average Screen Time
urban_rural_mean = df.groupby("Urban_or_Rural")["Avg_Daily_Screen_Time_hr"].mean()
print("\nAverage Screen Time by Locality:")
print(urban_rural_mean)

plt.figure(figsize=(6,4))
sns.barplot(
    x=urban_rural_mean.index,
    y=urban_rural_mean.values,
    hue=urban_rural_mean.index,
    palette="magma",
    legend=False
)
plt.title("Average Screen Time by Locality")
plt.xlabel("Locality")
plt.ylabel("Screen Time (hours/day)")
plt.show()
```

```python
# Device Usage Count Plot
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x="Primary_Device",
    hue="Primary_Device",
    palette="Set2",
    order=df["Primary_Device"].value_counts().index,
    legend=False
)
plt.title("Primary Screen Device Usage Count")
plt.xlabel("Device")
plt.ylabel("Number of Children")
plt.show()
```

```python
# Health Impact Counts
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    y="Health_Impacts",
    hue="Health_Impacts",
    order=df["Health_Impacts"].value_counts().index,
    palette="coolwarm",
    legend=False
)
plt.title("Health Impacts Reported")
plt.xlabel("Count")
plt.ylabel("Health Impact")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(
    data=df,
    x="Exceeded_Recommended_Limit",
    hue="Exceeded_Recommended_Limit",
    palette="Set1",
    legend=False
)
plt.title("Children Exceeding Recommended Screen Time Limits")
plt.xlabel("Exceeded Recommended Limit")
plt.ylabel("Count")
plt.show()
```
